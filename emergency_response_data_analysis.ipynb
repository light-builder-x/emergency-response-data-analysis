<h4>Check Data for null values and Replace</h4>

```python
# Count rows before removing nulls
rows_before = len(covid_df)
print(f"Rows before: {rows_before}")
```

```python
# Replace nulls with different values based on data type
df_cleaned = covid_df.copy()

for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':  # String column
        df_cleaned[column] = df_cleaned[column].fillna('-')
        print(f"Replaced nulls in string column '{column}' with '-'")
    elif df_cleaned[column].dtype in ['int64', 'float64', 'int32', 'float32']:  # Numeric column
        df_cleaned[column] = df_cleaned[column].fillna(0)
        print(f"Replaced nulls in numeric column '{column}' with 0")

print("\n\nCleaned DataFrame:")
print(df_cleaned)
```

```python
# Use the info function to check to see if there are any null values within the data set
covid_df = df_cleaned.copy()
covid_df.info()
```

```python
!pip install statsmodels
import statsmodels.api as sm

# First filter all data to lockdown period
lockdown_data = covid_df[(covid_df['created_date'] >= '2020-03-22 08:00:00') & (covid_df['created_date'] <= '2020-06-22 08:00:00')]

# Filter lockdown data for each agency
nypd_data = lockdown_data[lockdown_data['agency'] == 'NYPD']
dsny_data = lockdown_data[lockdown_data['agency'] == 'DSNY']
hpd_data = lockdown_data[lockdown_data['agency'] == 'HPD']
dot_data = lockdown_data[lockdown_data['agency'] == 'DOT']
dep_data = lockdown_data[lockdown_data['agency'] == 'DEP']

# Count requests by daily or weekly periods during lockdown
nypd_counts = nypd_data['created_date'].dt.to_period('D').value_counts().sort_index()
dsny_counts = dsny_data['created_date'].dt.to_period('D').value_counts().sort_index()
hpd_counts = hpd_data['created_date'].dt.to_period('D').value_counts().sort_index()
dot_counts = dot_data['created_date'].dt.to_period('D').value_counts().sort_index()
dep_counts = dep_data['created_date'].dt.to_period('D').value_counts().sort_index()

# Create aligned series for regression
df_regression = pd.DataFrame({
    'NYPD': nypd_counts,
    'DSNY': dsny_counts,
    'HPD': hpd_counts,
    'DOT': dot_counts,
    'DEP': dep_counts
}).fillna(0)

# Set up regression
y = df_regression['NYPD']
x = df_regression[['DSNY', 'HPD', 'DOT', 'DEP']]
x = sm.add_constant(x)  # Add intercept

model = sm.OLS(y, x, missing='drop')
result = model.fit()
print(result.summary())
```

<h4>How did the number of service requests change before and after the first lockdown?</h4>

<h5>Background</h5>

NYC First COVID Lockdown:
- **Start Date**: March 22, 2020 (8 PM)
- **End Date**: June 22, 2020

<h5>Date Selection Context</h5>

Based on source: https://citylimits.org/following-orders-a-timeline-of-state-and-city-action-on-covid/
**Start Date**: this was when New York State's "PAUSE" order (Policies that Assure Uniform Safety for Everyone) went into effect, which applied to NYC.

**End Date**: NYC didn't have a single "end date" but rather a phased reopening:
- The PAUSE order was extended through May 28 for New York City
- NYC began its initial reopening (Phase 1) on June 8, 2020, but later moved it to ***June 22, 2020***
- Phase 2 began on June 22, 2020
- So NYC's first lockdown lasted approximately 10-11 weeks, from March 22 through early June 22, 2020, with the city beginning a gradual, phased reopening rather than ending restrictions all at once.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Create figure with subplots
fig = plt.figure(figsize=(18, 12))

# ==================== Plot 1: Service Requests by Time Period ====================
ax1 = plt.subplot(2, 2, 1)
before_count = len(covid_df[covid_df['created_date'] < '2020-03-22 08:00:00'])
during_count = len(covid_df[(covid_df['created_date'] >= '2020-03-22 08:00:00') & (covid_df['created_date'] <= '2020-06-22 08:00:00')])
after_count = len(covid_df[covid_df['created_date'] > '2020-06-22 08:00:00'])

periods = ['Before\nCOVID', 'During\nLockdown', 'After\nReopening']
counts = [before_count, during_count, after_count]

bars = ax1.bar(periods, counts, color=['teal', 'purple', 'gold'], edgecolor='black', alpha=0.8)
ax1.set_title('Service Requests by Time Period', fontsize=13, fontweight='bold')
ax1.set_ylabel('Number of Requests', fontsize=11)
ax1.grid(axis='y', linestyle='--', alpha=0.4)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height):,}',
            ha='center', va='bottom', fontsize=10)

# ==================== Plot 2: Top Complaint Types Across Periods ====================
ax2 = plt.subplot(2, 2, 2)

before_complaints = covid_df[covid_df['created_date'] < '2020-03-22 08:00:00']['complaint_type'].value_counts().head(5)
during_complaints = covid_df[(covid_df['created_date'] >= '2020-03-22 08:00:00') & (covid_df['created_date'] <= '2020-06-22 08:00:00')]['complaint_type'].value_counts().head(5)
after_complaints = covid_df[covid_df['created_date'] > '2020-06-22 08:00:00']['complaint_type'].value_counts().head(5)

all_types = set(before_complaints.index) | set(during_complaints.index) | set(after_complaints.index)
periods_line = ['Before', 'During', 'After']

for complaint_type in all_types:
    counts_line = [
        before_complaints.get(complaint_type, 0),
        during_complaints.get(complaint_type, 0),
        after_complaints.get(complaint_type, 0)
    ]
    ax2.plot(periods_line, counts_line, marker='o', linewidth=2, markersize=8, label=complaint_type)

ax2.set_title('Top Complaint Types Across COVID Periods', fontsize=13, fontweight='bold')
ax2.set_xlabel('Time Period', fontsize=11)
ax2.set_ylabel('Number of Complaints', fontsize=11)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax2.grid(alpha=0.3)

# ==================== Plot 3: Borough Complaints Across Periods ====================
ax3 = plt.subplot(2, 2, 3)

before_boroughs = covid_df[covid_df['created_date'] < '2020-03-22 08:00:00']['borough'].value_counts().head(5)
during_boroughs = covid_df[(covid_df['created_date'] >= '2020-03-22 08:00:00') & (covid_df['created_date'] <= '2020-06-22 08:00:00')]['borough'].value_counts().head(5)
after_boroughs = covid_df[covid_df['created_date'] > '2020-06-22 08:00:00']['borough'].value_counts().head(5)

all_boroughs = set(before_boroughs.index) | set(during_boroughs.index) | set(after_boroughs.index)

for borough in all_boroughs:
    counts_boro = [
        before_boroughs.get(borough, 0),
        during_boroughs.get(borough, 0),
        after_boroughs.get(borough, 0)
    ]
    ax3.plot(periods_line, counts_boro, marker='s', linewidth=2, markersize=8, label=borough)

ax3.set_title('Borough Complaints Across COVID Periods', fontsize=13, fontweight='bold')
ax3.set_xlabel('Time Period', fontsize=11)
ax3.set_ylabel('Number of Complaints', fontsize=11)
ax3.legend(fontsize=10)
ax3.grid(alpha=0.3)

# ==================== Plot 4: Agency Service Requests Across Periods ====================
ax4 = plt.subplot(2, 2, 4)

before_agency = covid_df[covid_df['created_date'] < '2020-03-22 08:00:00']['agency_name'].value_counts().head(5)
during_agency = covid_df[(covid_df['created_date'] >= '2020-03-22 08:00:00') & (covid_df['created_date'] <= '2020-06-22 08:00:00')]['agency_name'].value_counts().head(5)
after_agency = covid_df[covid_df['created_date'] > '2020-06-22 08:00:00']['agency_name'].value_counts().head(5)

all_agencies = set(before_agency.index) | set(during_agency.index) | set(after_agency.index)

for agency in all_agencies:
    counts_agency = [
        before_agency.get(agency, 0),
        during_agency.get(agency, 0),
        after_agency.get(agency, 0)
    ]
    ax4.plot(periods_line, counts_agency, marker='D', linewidth=2, markersize=8, label=agency)

ax4.set_title('Agency Service Requests Across COVID Periods', fontsize=13, fontweight='bold')
ax4.set_xlabel('Time Period', fontsize=11)
ax4.set_ylabel('Number of Requests', fontsize=11)
ax4.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax4.grid(alpha=0.3)

# Add main title and adjust layout
plt.suptitle('NYC 311 Service Requests: COVID-19 Impact Analysis',
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.subplots_adjust(top=0.96)
plt.show()
```
